```python
import mne
# create a plot of annotation descriptions over time
annotationsT = rawA01T.annotations.description
annotationsE = rawA01E.annotations.description

# Annotation conversion map
annotation_map = {
    '276': 'Idling EEG (eyes open)',
    '277': 'Idling EEG (eyes closed)',
    '768': 'Start of a trial',
    '769': 'Cue onset left (class 1)',
    '770': 'Cue onset right (class 2)',
    '771': 'Cue onset foot (class 3)',
    '772': 'Cue onset tongue (class 4)',
    '783': 'Cue unknown',
    '1023': 'Rejected trial',
    '1072': 'Eye movements',
    '32766': 'Start of a new run',
}

# Map the annotations to their descriptions
annotationsT = [annotation_map.get(desc, desc) for desc in annotationsT]
annotationsE = [annotation_map.get(desc, desc) for desc in annotationsE]

# Plot the raw data
print("Annotations in A01T:", annotationsT)
print("Annotations in A01E:", annotationsE)

eventsT, event_idT = mne.events_from_annotations(rawA01T)
# Replace event IDs with annotation descriptions
event_idT = {annotation_map.get(str(key), str(key)): value for key, value in event_idT.items()}
mne.viz.plot_events(eventsT, rawA01T.info["sfreq"], event_id=event_idT, show=True)

eventsE, event_idE = mne.events_from_annotations(rawA01E)
event_idE = {annotation_map.get(str(key), str(key)): value for key, value in event_idE.items()}
mne.viz.plot_events(eventsE, rawA01E.info["sfreq"], event_id=event_idE, show=True)
```

```python
# use Morlet wavelets to calculate the difference in time-frequency power between classes
# allow you to see if the mu (812 Hz) or beta (1826 Hz) bands are actually desynchronizing during the imagined movements.
import numpy as np
from mne.time_frequency import tfr_morlet
freqs = np.arange(7, 30, 1)  # Frequencies from 7 to 30 Hz
n_cycles = freqs / 2.  # Different number of cycle per frequency
# Check available channels and pick motor and parietal channels
available_channels = rawA01T.ch_names
motor_channels = [ch for ch in available_channels if ch in ['C3', 'Cz', 'C4', 'P3', 'Pz', 'P4']]
picks = mne.pick_channels(available_channels, motor_channels)
# Create epochs for left and right hand motor imagery
epochs_left = mne.Epochs(rawA01T, eventsT, event_id={'Cue onset left (class 1)': 769}, tmin=0, tmax=4, picks=picks, baseline=None, preload=True)
epochs_right = mne.Epochs(rawA01T, eventsT, event_id={'Cue onset right (class 2)': 770}, tmin=0, tmax=4, picks=picks, baseline=None, preload=True)
# Compute TFR for left and right hand motor imagery
tfr_left = tfr_morlet(epochs_left, freqs=freqs, n_cycles=n_cycles, return_itc=False, average=True)
tfr_right = tfr_morlet(epochs_right, freqs=freqs, n_cycles=n_cycles, return_itc=False, average=True)
# Compute the difference in power between left and right hand motor imagery
tfr_diff = tfr_left.copy()
tfr_diff.data = tfr_left.data - tfr_right.data
# Plot the difference in TFR
tfr_diff.plot(picks=['C3', 'C4'], title='TFR Difference (Left - Right)', show=True) 
```

```python
import matplotlib.pyplot as plt
from mne.decoding import CSP
from pyriemann.estimation import Covariances
from pyriemann.tangentspace import TangentSpace
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis as LDA
from sklearn.linear_model import LogisticRegression
from sklearn.pipeline import make_pipeline

import moabb
import moabb.analysis.plotting as moabb_plt
from moabb.analysis.meta_analysis import (  # noqa: E501
    compute_dataset_statistics,
    find_significant_differences,
)
from moabb.datasets import BNCI2014_001
from moabb.evaluations import CrossSessionEvaluation
from moabb.paradigms import LeftRightImagery, MotorImagery


moabb.set_log_level("info")

print(__doc__)
```

```python
pipelines = {}

pipelines["CSP+LDA"] = make_pipeline(CSP(n_components=8), LDA())

pipelines["RG+LR"] = make_pipeline(Covariances(), TangentSpace(), LogisticRegression())

pipelines["CSP+LR"] = make_pipeline(CSP(n_components=8), LogisticRegression())

pipelines["RG+LDA"] = make_pipeline(Covariances(), TangentSpace(), LDA())
```

```python
paradigm = LeftRightImagery()
dataset = BNCI2014_001()
dataset.subject_list = dataset.subject_list[:4]
datasets = [dataset]
overwrite = True  # set to False if we want to use cached results
evaluation = CrossSessionEvaluation(
    paradigm=paradigm, datasets=datasets, suffix="stats", overwrite=overwrite
)

results = evaluation.process(pipelines)
```

```python
fig = moabb_plt.score_plot(results)
plt.show()
```

```python
fig = moabb_plt.paired_plot(results, "CSP+LDA", "RG+LDA")
plt.show()
```

```python
stats = compute_dataset_statistics(results)
P, T = find_significant_differences(stats)
```

```python
fig = moabb_plt.meta_analysis_plot(stats, "CSP+LDA", "RG+LDA")
plt.show()
```

```python
moabb_plt.summary_plot(P, T)
plt.show()
```
